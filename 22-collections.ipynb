### ChainMap
    dict-like class for creating a single view of multiple mappingss

```python
from collections import ChainMap 
```

```python
dict_1 = {'a': 1, 'b': 2} 
dict_2 = {'c': 3, 'd': 4} 
dict_3 = {'e': 5, 'f': 6} 
```

```python
# Defining the chainmap  
result = ChainMap(dict_1, dict_2, dict_3)
result
```

```python
result.maps
```

```python
list(result.keys())
```

```python
dict_4 = {'s': 6, 'b': 0}
```

```python
res = result.new_child(dict_4)
res
```

```python
res.maps.append({'m': 13})
res
```

```python
for mapping in res.maps:
    print(mapping)
```

```python
for key, value in res.items():
    print(key,'---', value)
    
```

```python
res['b']
```

```python
res.maps = reversed(res.maps)
res
```

```python
res['b']
```

### Counter
    dict subclass for counting hashable objects

```python
word = "geeksforgeeks"
counter = {}
for letter in word:
    counter[letter] = counter.get(letter, 0) + 1


counter
```

```python
from collections import Counter
```

```python
example = Counter("geeksforgeeks")
example
```

```python
for item in example.elements():
    print (item, end = " ")
```

```python
example_2 = Counter({'geeks' : 4, 'for' : 1, 'gfg' : 2, 'python' : 3})
example_2
```

```python
for item in example_2.elements():
    print ( item, end = " ")
```

```python
letters = Counter({"i": 4, "s": 4, "p": 2, "m": 1})
letters.update("missouri")
letters
```

### defaultdict()
    dict subclass that calls a factory function to supply missing values 

```python
dic = {}
dic["a"] = 1
dic["b"] = 2
  
print(dic["a"])
print(dic["b"])
print(dic["c"])
```

```python
from collections import defaultdict
```

```python
dic = defaultdict(int)
dic["a"] = 1
dic["b"] = 2
  
print(dic["a"])
print(dic["b"])
print(dic["c"])
```

```python
dic = defaultdict(list)
dic["c"]
```

```python
dic = defaultdict(set)
dic["c"]
```

```python
dic = defaultdict(list)
  
for num in range(5):
    dic[num].append(num)
```

```python
dic
```

```python
dic = defaultdict(int)
   
number = [1, 2, 3, 4, 2, 4, 1, 2]
for num in number:
    dic[num] += 1
dic
```
